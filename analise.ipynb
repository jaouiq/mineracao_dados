Nesse exemplo, vamos analisar o _dataset_ da taxa de cancelamentos de planos dos usurios da empresa farmacutica **Umbrella Corporation.**


### Trazendo os dados do _dataset_ para o notebook:

```python
# importa a biblioteca panda
import pandas as pd

# l os dados do dataset
tabela = pd.read_csv('cancelamentos.csv')

# mostra os dados da tabela na tela
display(tabela)
```

Agora, vou eliminar da consulta dados desnecessrios. Vou eliminar o ID do usurio.

```python
tabela = tabela.drop('CustomerID', axis=1)
display(tabela)
```

Vamos descobrir quantas linhas com informaes vazias ns temos.

```python
display(tabela.info())
```

Vamos eliminar as clulas vazias da nossa consulta.

```python
tabela = tabela.dropna()
display(tabela.info())
```

Vamos contar quantos usurios cancelaram e quantos no cancelaram.

```python
display(tabela['cancelou'].value_counts())
```

Traduzindo para porcentagem:

```python
display(tabela['cancelou'].value_counts(normalize=True).map('{:.2%}'.format))
```

Prxima anlise de dados vai ser para verificar a durao dos contratos.

```python
display(tabela['duracao_contrato'].value_counts())
display(tabela['duracao_contrato'].value_counts(normalize=True).map('{:.2%}'.format))
```

Vamos agrupar as informaes da durao do contrato:

```python
display(tabela.groupby('duracao_contrato').mean(numeric_only=True))
```

Vamos eliminar os dados dos contratos mnesais.

```python
tabela = tabela[tabela['duracao_contrato'] != 'Monthly']
display(tabela)
display(tabela['cancelou'].value_counts())
display(tabela['cancelou'].value_counts(normalize=True).map('{:.2%}'.format))
```
